# BMR and Daily Energy Expenditure calculator
The subject inserts his/her sex, height (in cm)) and weight (in kg). From there it is possible to calculate the BMI and BMR. 

If the BMI is lower than 19 (limit of underweight), the BMR is increased (around 10%) to help to reach the state of normal weight (the optimal status). 
If the BMI is higher than 29.0 the subject is overweight. If it due to a high muscle mass the BMR does not change; on the other hand, if this situation is due to an excess of fat mass (it is asked), the BMR adapted (reduced of 15%) as fat mass does not contribute to the metabolism.

After the BMR is calculated and adapted to the body status, the subject is asked the PAL and from there it possible to know the Daily Energy Expenditure.



```python
while True:
        "Calcolo BMR"
        sesso = input("insert sex , m for male or f for female  ")
        altezza = float(input ("insert height in cm  "))
        peso = float (input ("insert weight in Kg  "))
        età = float(input ("insert age in years "))
        BMI= (peso)/((altezza/100)*(altezza/100))
        print ("BMI "+ str((round(BMI, 2))))
        if sesso == "m" and 19 <= BMI <= 29.9 :
            BMR = (66.5 + (peso * 13.75) + (altezza * 5.003) - (6.755 * età)) 
            status = "Status: normal weight"
            print (status)
            print ("BMR " + (str(int(BMR))) + " Kcal")
        elif sesso == "m" and BMI <= 18.9 :
            BMR = (66.5 + (peso * 13.75) + (altezza * 5.003) - (6.755 * età)) * 1.10
            status = "Status: underweight"
            print (status)
            print ("BMR adapted to underweight situation: " + (str(int(BMR))) + " Kcal")
        elif sesso == "m" and 30 <= BMI :
            question = input(" insert body shape: a) for overweight b) bodybuilder")
            BMR = 66.5 + (peso * 13.75) + (altezza * 5.003) - (6.755 * età)
            if question == "a":
                status = "Status: overweight"
                print (status)
                print (" BMR adapted to overwight situation")
                
                BMR = BMR * 0.85
                print ("BMR " + (str(int(BMR))) + " Kcal")
            elif question == "b":
                status = "Status: high muscle"
                print (status)
                print ("BMR " + (str(int(BMR))) + " Kcal")
                
        elif sesso == "f" and 19 <= BMI <= 29.9 :
            BMR = 655.1+ (peso * 9.563) + (altezza * 1.850) - (4.676 * età)
            status = "Status: normal weight"
            print (status)
            print ("BMR " + (str(int(BMR))) + " Kcal")
        elif sesso == "f" and BMI <= 18.9 :
            BMR = (655.1+ (peso * 9.563) + (altezza * 1.850) - (4.676 * età)) * 1.10
            status = "Status: underweight"
            print (status)
            print ("BMR adapted to underweight situation: " + (str(int(BMR))) + " Kcal")
        elif sesso == "f" and 30 <= BMI:
             question = input(" insert body shape: a) for overweight b) bodybuilder")
             BMR = 655.1+ (peso * 9.563) + (altezza * 1.850) - (4.676 * età)
             if question == "a":
                 status = "Status: overweight"
                 print (status)
                 print (" BMR adapted to overwight situation")
                 BMR = BMR * 0.85
                 print ("BMR " + (str(int(BMR))) + " Kcal")
             elif question == "b":
                 status = "Status: high muscle"
                 print (status)
                 print ("BMR " + (str(int(BMR))) + " Kcal")
             

        PAL = input("Indicate level of physical activity with: \n a)sedentary, office work, no exercise.:   \n b) Moderately active,construction worker or training one hour a day: \n c)Vigorously active agricultural worker or person who trains two hours a day \n d)Extremely active professional athlete: \n ")
        if PAL == "a":
                calorie = BMR * 1.55
                print("Daily Energy Expenditure " + str(int(calorie)) + " Kcal")
        
                 
        elif PAL == "b":
             calorie = BMR * 1.85
             print("Daily Energy Expenditure " + str(int(calorie)) + " Kcal") 

        elif PAL == "c":
             calorie = BMR * 2.20
             print("Daily Energy Expenditure " + str(int(calorie)) + " Kcal") 

        elif PAL == "d":
             calorie = BMR * 2.50
             print("Daily Energy Expenditure " + str(int(calorie)) + " Kcal") 


        fine = input ("Do you want to continue? y/n ")
        if fine == "y":
            continue
        elif fine == "n":
            break
```
